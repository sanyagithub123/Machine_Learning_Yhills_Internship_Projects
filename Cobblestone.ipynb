```python
import numpy as np
import matplotlib.pyplot as plt

def detect_anomalies(data, algorithm='moving_average', threshold=2.0):
    # Implement anomaly detection logic based on the chosen algorithm
    if algorithm == 'moving_average':
        # Example: Moving Average Anomaly Detection
        moving_avg = np.mean(data)
        std_dev = np.std(data)
        anomalies = np.abs(data - moving_avg) > threshold * std_dev
    elif algorithm == 'z_score':
        # Example: Z-Score Anomaly Detection
        z_scores = (data - np.mean(data)) / np.std(data)
        anomalies = np.abs(z_scores) > threshold
    # Add more algorithms as needed
    return anomalies

    # Simulated data stream
data_stream = np.random.normal(0, 1, size=1000)

# Continuously process data
for data_point in data_stream:
    # Add data point to the history
    # Apply anomaly detection function
    is_anomaly = detect_anomalies(data_point)

    if is_anomaly:
        print(f"Anomaly Detected: {data_point}")
    else:
        print(f"Normal Data: {data_point}")

        plt.plot(data_stream, label='Data Stream')
plt.scatter(np.where(is_anomaly)[0], data_stream[is_anomaly], color='red', label='Anomalies')
plt.legend()
plt.show()



```
